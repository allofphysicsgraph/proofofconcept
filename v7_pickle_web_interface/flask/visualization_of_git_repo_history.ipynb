# visualization of git history analytics

```python
import pandas
print('pandas',pandas.__version__)
import matplotlib.pyplot as plt
!hostname
```

See https://physicsderivationgraph.blogspot.com/2020/04/analyze-each-commit-in-repo.html

```python
with open('record.log','r') as fil:
    record_content = fil.readlines()
```

```python
list_of_dicts = []
for this_recrd in record_content:
    list_of_dicts.append(eval(this_recrd.strip()))
```

```python
df = pandas.DataFrame(list_of_dicts)
df.shape
```

```python
df['date as datetime'] = pandas.to_datetime(df['date'], infer_datetime_format=True)
```

```python
df.columns
```

```python
# https://stackoverflow.com/a/13873014/1164295
ax = df.plot(x='date as datetime', y='html file count', style=".", logy=True)
df.plot(x='date as datetime', y='python file count', style=".", ax=ax)
df.plot(x='date as datetime', y='png file count', style=".", ax=ax)
df.plot(x='date as datetime', y='sh file count', style=".", ax=ax)
df.plot(x='date as datetime', y='xml file count', style=".", ax=ax)
plt.xlabel('date')
plt.ylabel('file count');
plt.legend(bbox_to_anchor=(1.05, 1));
```

```python
ax = df.plot(x='date as datetime', y='html line count', style=".", logy=True)
df.plot(x='date as datetime', y='python line count', style=".", ax=ax)
df.plot(x='date as datetime', y='tex line count', style=".", ax=ax)
df.plot(x='date as datetime', y='sh line count', style=".", ax=ax)
df.plot(x='date as datetime', y='xml line count', style=".", ax=ax)
plt.xlabel('date')
plt.ylabel('file count');
plt.legend(bbox_to_anchor=(1.05, 1))
```
