```python
from scipy.stats import powerlognorm
import matplotlib.pyplot as plt
import numpy as np
print('numpy',np.__version__)
```

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.powerlognorm.html#scipy.stats.powerlognorm

```python
c, s = 2.14, 0.446
mean, var, skew, kurt = powerlognorm.stats(c, s, moments='mvsk')
```

```python
x = np.linspace(powerlognorm.ppf(0.01, c, s),
                powerlognorm.ppf(0.99, c, s), 100)
ax = plt.plot(x, powerlognorm.pdf(x, c, s),
       'r-',label='powerlognorm pdf')
```

```python
r = powerlognorm.rvs(c, s, size=100000)
```

```python
_=plt.hist(r, 30, density=True)
```

```python
_=plt.hist(np.log(r), 30, density=True)
```

```python
_=plt.hist(np.log(np.log(r)+5), 30, density=True)
```
